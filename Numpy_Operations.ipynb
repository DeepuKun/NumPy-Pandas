```python
#Deep and Shallow COPY
```

```python
a = np.arange(1,21)
a
```

```python
#When we slice an Numpy Array and save it into a variable, then if we make changes to that sliced variable,
#it does not chnage anything in the Original Array
```

```python
#Hence in Numpy Array, Deep Copies are made, while in lists, Shallow Copies are made
```

```python
b = a
b[0]= 99
b
```

```python
a
```

```python
# if we do b = a then they both point to the same part in the memory, hence a shallow copy 
```

```python
c = a.copy()
```

```python
c[0] = 1
c
```

```python
a
```

```python
# if we use .copy() function, it makes a deep copy
```
