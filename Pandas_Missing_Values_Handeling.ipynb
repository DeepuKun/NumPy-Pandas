```python
import numpy as np
import pandas as pd
```

```python

data = {
    'X': [10, 20, np.nan, 40, 50],
    'Y': [np.nan, 25, 35, 45, 55],
    'Z': [5, 15, 25, np.nan, np.nan],
    'W': [np.nan, 12, np.nan, 42, 52]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.isna()
```

```python
 df.isna().sum()
```

```python
df.isna().any() #to check if ANY null value is there in that COL
```

```python
df.dropna() #drop all the rows with null values
```

```python
df
```

```python
df.dropna(thresh=3) #if the row has more that or equal to 3 non null values, keep it
```

```python
#FILLING NULL VALUES
df.fillna(0)
```

```python
values = {'X':10, 'Y':20, 'Z':30, 'W':40}
df.fillna(value=values) #Add Coloumn wise Specific values to the NULL VALUES
```

```python
df.fillna(df.mean())
```
